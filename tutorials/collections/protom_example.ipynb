# Datamine PROTOM Process Example

This notebook demonstrates how to create a block model prototype using the **`protom`** command wrapper in `dmstudio`.

The prototype defines the coordinates limits, cell sizes, and dimensions of the block model that will subsequently be used for geological modeling or grade estimation.

```python
# Step 1: Connect to Datamine and Initialize Sandbox
import os
import glob
from dmstudio import dmfiles, initialize, agent

# Connect to running Studio RM instance
dm_fil = dmfiles.init(version='StudioRM')
oScript = initialize.studio('StudioRM')
print(f"Connected to: {oScript.Caption}")

# Initialize active project sandbox using the shared test_sandbox project
notebook_folder = os.path.normpath(os.path.dirname(os.path.abspath('__file__'))).lower()
agent.initialize_sandbox(notebook_folder)
```

## Step 2: Define and Generate the Block Model Prototype
The `protom` command requires interactive inputs. We pass these values (Mined-out field: No, Subcells: Yes, Origin: 5900, 4800, -100, Cell size: 10,10,10, Counts: 25,45,32) via the `arguments` parameter.
The output will be created in this notebook's dedicated directory with a temporary `t_` prefix.

```python
# Run PROTOM to create 't_model_proto.dmx' in active sandbox
proto_path = 't_model_proto'

print("Generating model prototype...")
dm_fil.protom(
    out_o=proto_path,
    rotmod_p=0,
    arguments=" 'N' 'Y' '5900' '4800' '-100' '10' '10' '10' '25' '45' '32'"
)
print(f"Prototype created: {proto_path}.dmx")
```

## Step 3: Clean up Project Folder
Run this cell to delete all generated files starting with `t_` and restore the directory to its clean state.

```python
# Clean up all temporary files matching t_*.*
temp_files = glob.glob("t_*.*")
for f in temp_files:
    os.remove(f)
    print(f"Cleaned up: {os.path.basename(f)}")
```
